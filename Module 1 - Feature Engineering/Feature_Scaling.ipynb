## Standardization

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
# lets standardize the total bill feature
x = list(df['total_bill'])
```

```python
import numpy as np
mu = round(np.mean(x),3)
sd = round(np.std(x), 3)
mu, sd
```

```python
standardized = []
for x_i in x:
    z = (x_i - mu)/sd
    standardized.append(z)

# standardized
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1, 2, 1)    #1 row 2 column 1st plt
plt.hist(x, histtype="bar")
plt.title("Before Standardization")
plt.xlabel("X")


plt.subplot(1, 2, 2)    #1 row 2 column 2nd plt
plt.hist(standardized)
plt.title("After Standardization")
plt.xlabel("X_Standardized")

plt.show()
# see the x values.
```

```python
# We can do same using sklearn
# https://scikit-learn.org/1.6/modules/generated/sklearn.preprocessing.StandardScaler.html 

# class sklearn.preprocessing.StandardScaler(*, copy=True, with_mean=True, with_std=True)

from sklearn.preprocessing import StandardScaler
```

```python
# where u is the mean of the training samples or zero if with_mean=False, and s is the standard deviation of the training samples or one if with_std=False.

scaler = StandardScaler()   #creating an object
scaler
```

```python
scaler.fit(df[['total_bill']])  #expects a 2d 

# Compute the mean and std to be used for later scaling.
```

```python
import pandas as pd
total_bill_scaled = pd.DataFrame(scaler.transform(df[["total_bill"]]))
total_bill_scaled
```

## Normalization
- **Min-Max**:  $x_{scaled}$ = $ ({x_{i} - x_{min}}) / ({x_{max} - x_{min}})$

```python
df = sns.load_dataset('taxis')
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
# min_max.fit(df[['distance', 'fare', 'tip']])
```

```python
pd.DataFrame(min_max.fit_transform(df[['distance', 'fare', 'tip']]))
```

```python
# now if we want to add new data points we can use min_max.transform
min_max.transform([[1,2,3]])
```
