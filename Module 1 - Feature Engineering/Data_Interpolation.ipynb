## Linear Interpolation

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)

plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
# it's a linear plot
# for linear interpolation we can use numpy
x_new = np.linspace(1, 5, 10)   #creating 10 new x points
y_interp = np.interp(x= x_new, xp= x, fp= y)

y_interp
```

```python
plt.scatter(x_new, y_interp)
plt.xlabel("X_New")
plt.ylabel("Interpolated Y values")
plt.title("After Interpolation")
plt.show()
```

## Cubic Interpolation
we will use scipy here. We can do linear interpolation using this as well

```python
x = np.array([1,2,3,4,5])
y = x**3        #we are making a cubic data intentionally
```

```python
from scipy.interpolate import interp1d
```

```python
# we have to create a cubic interpolation object or function
f_cubic = interp1d(x = x, y = y, kind= "cubic")
f_cubic
```

```python
x_new
```

```python
y_cubic_interp = f_cubic(x_new)

y_cubic_interp
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1, 2, 1)    #1 row 2 column 1st plt
plt.scatter(x, y)
plt.title("Before interpolation")
plt.xlabel("X")


plt.subplot(1, 2, 2)    #1 row 2 column 2nd plt
plt.scatter(x_new, y_cubic_interp)
plt.title("After interpolation")
plt.xlabel("X_New")
```

## Polynomial Interpolation

```python
x
```

```python
y = 5*(x**2) + 9    #2 degree polynomial
```

```python
# scipy don't have a kind poly. we wil use numpy only
p = np.polyfit(x , y = y, deg = 2) #Least squares polynomial fit.   #creates a fitted polynomial function
```

```python
x_new
```

```python
y_p_interp = np.polyval(p, x_new)
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1, 2, 1)    #1 row 2 column 1st plt
plt.scatter(x, y)
plt.title("Before interpolation")
plt.xlabel("X")


plt.subplot(1, 2, 2)    #1 row 2 column 2nd plt
plt.scatter(x_new, y_p_interp)
plt.title("After interpolation")
plt.xlabel("X_New")
```
