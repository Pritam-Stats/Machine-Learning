### Iris dataset - Binary only

```python
## Importing all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris

data = load_iris()
data
```

```python
data.keys()
```

```python
## let's see the description of the dataset
print(data.DESCR)
```

Iris plants dataset
--------------------

**Data Set Characteristics:**

:Number of Instances: 150 (50 in each of three classes)
:Number of Attributes: 4 numeric, predictive attributes and the class

**:Attribute Information:**

- sepal length in cm
- sepal width in cm
- petal length in cm
- petal width in cm
- class:
    - Iris-Setosa
    - Iris-Versicolour
    - Iris-Virginica


### **In short there are 4 features and 3 classes. We have to predict the class based on the features. So it's a classification Problem.**

```python
data.feature_names
```

```python
## Target data, we can observe that 3 classes are already encoded in 0, 1, 2        
## And it's a absolutely balanced dataset
np.unique(data['target'], return_counts = True)
```

```python
## Target class names
data.target_names
```

## DataFrame

```python
df = pd.DataFrame(data = data.data, columns = data.feature_names)
df
```

```python
## adding the target variable to the datasets
df['Target'] = data.target
df
```

- In total there 150 observations.

### Now we will only take the binary classification, so we need to extract the data only for Target 0 and 1

```python
data_b = df[df['Target'] != 2]      # data where target is not equal to 2
data_b
```

```python
## Now the new dataset there are 100 observations   # so 50 observations excluded. we can verify this
(data.target == 2).sum()
```

### so 'data_b' is our working dataset - Now let's define X and y

```python
X = data_b.iloc[ : , : -1]
y = data_b.iloc[ : , -1]
```

```python
X
```

```python
y
```

## train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 95)
```

```python
X_train
```

## Model

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()
```

```python
logistic_model.fit(X = X_train, y = y_train)
```

```python
logistic_model.intercept_
```

```python
logistic_model.coef_
```

## Logistic Model :

# $h_{\theta}(X) = \frac{1}{1 + \frac{1}{\exp{(\theta_{0} + \theta_{1}{X_1} + \dots + \theta{n}{X_n})}}}$

## $\theta_0 = -6.71541029$
## $(\theta_1 , \theta_2 , \theta_3 , \theta_4) = ([ 0.45891181, -0.78594362,  2.17433296,  0.87080449])$

```python
y_pred_test = logistic_model.predict(X_test)
y_pred_test
```

## Evaluation Metrics

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_true=y_test, y_pred= y_pred_test)
cm
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels= logistic_model.classes_)
cm_display.plot(cmap = "Blues")
plt.show()
```

#### **TP = 12, TN = 13, FP = 0, FN = 0**
- Means No false prediction - Means accuracy is 100%

```python
## Accuracy Score
accuracy_score(y_true=y_test, y_pred=y_pred_test)
```

```python
print(classification_report(y_true= y_test, y_pred= y_pred_test))
```

```python
## Another important thing in classification problems is prediction probabilities

pd.DataFrame(logistic_model.predict_proba(X = X_test), columns = ['Pr(y = 0)', 'Pr(y = 1)'])
```

# Hyper-Parameter Tuning

### **GridSearchCV**
### **RandomizedSearchCV**

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

## for grid search cv we have to give estimator which our model object and param_grid we have specify some parameters of logistic_regression parameters in a dict to cv
#### this is to find out the best parameter combination to make the model accuracy better. (Though here our accuracy is already 1, so won't give much effect)
- **GridSearchCV** = https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html#sklearn.model_selection.GridSearchCV
- **Logistic Regression** = https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
parameters = {
    'penalty' : ('l1', 'l2', 'elasticnet', None),
    'C' : [1, 5, 9],
    'solver' : ('lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga')
}

# we could have also given more parameters like solver and more
gridCV = GridSearchCV(estimator = LogisticRegression(), param_grid = parameters, cv = 5)
```

```python
## In this step the train data get's further splitted to train and validation data as we have give cv = 5
gridCV.fit(X_train, y_train)
```

```python
### To find out the best parameter combination
gridCV.best_params_
```

### Now we specify this parameter while training our logistic model

```python
model_after_gridCV = LogisticRegression(penalty= 'l1', C = 1,  solver = 'liblinear')
model_after_gridCV.fit(X=X_train, y=y_train)
```

```python
model_after_gridCV.intercept_
```

```python
logistic_model.intercept_
```

```python
model_after_gridCV.coef_
```

```python
logistic_model.coef_
```

- **Here intercept and some coefficients became 0, this is because of the l1 regularization and liblinear optimizations. L1 or Lasso regularization is used for 'Feature Selection', it pushes those coefficients to zero which are not very significant, Since the model was already performing perfectly (accuracy = 1) before tuning accuracy remained the same after tuning, it means the removed features did not contribute much to classification.**

- **This suggests feature redundancy, meaning the remaining two nonzero features were enough to perfectly separate class 0 and class 1.**

```python
y_pred2_test = model_after_gridCV.predict(X_test)
```

### Confusion Matrix after GridSearchCV

```python
cm2 = confusion_matrix(y_true=y_test, y_pred=y_pred2_test)
cm2_display = ConfusionMatrixDisplay(confusion_matrix = cm2, display_labels= model_after_gridCV.classes_)
cm_display.plot(cmap = "Blues")
plt.show()
```

```python
print(classification_report(y_true=y_test, y_pred= y_pred2_test))
```

```python
import pandas as pd
import seaborn as sns
# Extract feature names and coefficients
feature_names = data.feature_names
coefficients = model_after_gridCV.coef_[0]

# Create DataFrame for visualization
coef_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
coef_df = coef_df.sort_values(by="Coefficient", ascending=False)

# Plot coefficients
plt.figure(figsize=(9, 5))
sns.barplot(x="Coefficient", y="Feature", data=coef_df)
plt.title("Feature Importance (Logistic Regression Coefficients) after GridSearchCV")
plt.axvline(0, color='gray', linestyle='dashed')
plt.show()
```

```python
### Randomized search cv
from sklearn.model_selection import RandomizedSearchCV
random_cv = RandomizedSearchCV(estimator=LogisticRegression(), param_distributions=parameters, cv = 5, n_iter = 10) #n_iter is 10 at default
random_cv.fit(X = X_train, y=y_train)
```

```python
random_cv.best_params_
```

```python
model_after_randomCV = LogisticRegression(C=9, penalty=None, solver='saga')
model_after_randomCV.fit(X=X_train, y = y_train)
```

```python
model_after_randomCV.coef_
```

```python
model_after_randomCV.intercept_
```

```python
import pandas as pd
import seaborn as sns
# Extract feature names and coefficients
feature_names = data.feature_names
coefficients_random = model_after_randomCV.coef_[0]

# Create DataFrame for visualization
coef_random_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients_random})
coef_random_df = coef_random_df.sort_values(by="Coefficient", ascending=False)

# Plot coefficients
plt.figure(figsize=(9, 5))
sns.barplot(x="Coefficient", y="Feature", data=coef_random_df)
plt.title("Feature Importance (Logistic Regression Coefficients) After RandomizedSearchCV")
plt.axvline(0, color='red', linestyle='dashed')
plt.show()
```

```python
cm3 = confusion_matrix(y_true=y_test, y_pred=y_pred3_test)
```

```python
cm3_display = ConfusionMatrixDisplay(confusion_matrix = cm3, display_labels = model_after_randomCV.classes_)
cm3_display.plot(cmap= "GnBu")
plt.show()
```
